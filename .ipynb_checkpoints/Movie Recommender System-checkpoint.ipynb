```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('Netflix_Titles.csv')
```

```python
df.head(5)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df2 = df.drop(['date_added'],axis='columns')
```

```python
df2.sample(5)
```

```python
df2['title'] = df2['title'].fillna('other')
```

```python
df2['director'] = df2['director'].fillna('other')
df2['cast'] = df2['cast'].fillna('other')
df2['country'] = df2['country'].fillna('other')
df2['director'] = df2['director'].fillna('other')
```

```python
df2['rating'] = df2['rating'].fillna('other')
```

```python
df2['duration'] = df2['duration'].fillna('other')
```

```python
df2.head(1)
```

```python
def remove_space(L):
    L1 = L.split(" ")
    return "".join(L1)
```

```python
df2['type'] = df2['type'].apply(remove_space)
df2['director'] = df2['director'].apply(remove_space)
df2['country'] = df2['country'].apply(remove_space)
df2['duration'] = df2['duration'].apply(remove_space)
df2['listed_in'] = df2['listed_in'].apply(remove_space)
```

```python
df2
```

```python
df2['release_year'] = df2["release_year"].map(str)
```

```python
df2['tags'] = df2['type']+ ','+df2['director']+ ','+df2['cast']+ ','+df2['country']+ ','+df2['release_year']+ ','+df2['rating']+ ','+df2['duration']+ ','+df2['listed_in']
df2.sample(4)
```

```python
new = df2[['show_id','title','tags']]
```

```python
new.head(5)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 8007,
                    stop_words = 'english')
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(f'{i} movie is',endl="")
        print(new.iloc[i[0]].title)
```

```python
recommend('Ganglands')
```
