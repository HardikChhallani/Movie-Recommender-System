```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('movies_data_1.csv')
df
```

```python
df2 = df.drop(['date_added'],axis='columns')
```

```python
df2['title'] = df2['title'].fillna('other')
```

```python
df2['director'] = df2['director'].fillna('other')
df2['cast'] = df2['cast'].fillna('other')
df2['country'] = df2['country'].fillna('other')
df2['director'] = df2['director'].fillna('other')
df2['rating'] = df2['rating'].fillna('other')
df2['duration'] = df2['duration'].fillna('other')
```

```python
def remove_space(L):
    L1 = L.split(" ")
    return "".join(L1)
```

```python
df2['type'] = df2['type'].apply(remove_space)
df2['director'] = df2['director'].apply(remove_space)
df2['country'] = df2['country'].apply(remove_space)
df2['duration'] = df2['duration'].apply(remove_space)
df2['listed_in'] = df2['listed_in'].apply(remove_space)
df2['release_year'] = df2["release_year"].map(str)
df2
```

```python
df2['tags'] = df2['type']+ ' '+df2['director']+ ' '+df2['cast']+ ' '+df2['country']+ ' '+df2['release_year']+ ' '+df2['rating']+ ' '+df2['duration']+ ' '+df2['listed_in']+" "+df2['description']
```

```python
new = df2[['show_id','title','tags']]
new
df2
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
new['tags'] = new['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 8807,stop_words = 'english')
vector = cv.fit_transform(new['tags']).toarray()
```

```python
len(cv.get_feature_names_out())
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
```

```python
def recommend(movie):
    movie_index = new[new['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key = lambda x : x[1])
    recommend_movies = []
    for i in movies_list[1:6]:
        recommend_movies.append(new.iloc[i[0]].title)
        
    return recommend_movies
```

```python
movies_returned = recommend('Same Kind of Different as Me')
movies_returned
```
